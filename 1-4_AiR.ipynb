Teraz łączymy poszczególne kroki, aby ich wyniki przekazać dalej

```python
rent[["id", "price","flat_area", "quarter"]].query(
    "price>=1000 & price<=10000 & flat_area<=200").groupby(
        "quarter").agg(
            price_med = pd.NamedAgg("price", "median"), 
            flat_med = pd.NamedAgg("flat_area", "median"),
            count = pd.NamedAgg("price", "count")).sort_values(
                by="count", ascending = False)
```

```python
(
rent[["id", "price","flat_area", "quarter"]]
  .query( "price>=1000 & price<=10000 & flat_area<=200")
  .groupby("quarter")
  .agg(
    price_med = pd.NamedAgg("price", "median"), 
    flat_med = pd.NamedAgg("flat_area", "median"),
    count = pd.NamedAgg("price", "count"))
  .sort_values(by="count", ascending = False)
)
```

```python
(
rent[["id", "price","flat_area", "quarter"]]
  .query( "price>=1000 & price<=10000 & flat_area<=200")
  .groupby("quarter")
  .agg(
    price_med = pd.NamedAgg("price", "median"), 
    flat_med = pd.NamedAgg("flat_area", "median"),
    count = pd.NamedAgg("price", "count"))
  .sort_values(by="count", ascending = False)
  .assign(a=1) ##dodanie kolumny
)
```

```python
pd.set_option("max_rows", 100) ##zmienia liczbe wierszy, która jest pokazywana w wyniku
(
rent[["id", "price","flat_area", "quarter"]]
  .query( "price>=1000 & price<=10000 & flat_area<=200")
  .groupby("quarter")
  .agg(
    price_med = pd.NamedAgg("price", "median"), 
    flat_med = pd.NamedAgg("flat_area", "median"),
    count = pd.NamedAgg("price", "count"))
  .sort_values(by="count", ascending = False)
  .assign(a=1) ##dodanie kolumny
)
```
