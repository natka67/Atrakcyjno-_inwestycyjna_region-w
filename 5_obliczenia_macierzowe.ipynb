```python
A = np.array([(1, 2, -1, 0),(3,-2, 4, 5),(2, 6, 5,-3),(0, 1, 5, -4)])
B = np.array([(3, 6),(4,0),(2,-1),(1,1)])
```

```python
#1 Oblicz wyznacznik macierzy
np.linalg.det(A)
```

```python
#2 (3 sposoby) ZnajdÅº iloczyn A*B
A@B #tylko w pythonie w wersji 3 i wyÅ¼ej
np.matmul(A,B) #matmul = matrix multiplication
A.dot(B)
```

```python
#3 (3 sposoby) ZnajdÅº macierz transponowanÄ… B
B.T
B.transpose()
np.transpose(B)
```

```python
#4 ZnajdÅº macierz odwrotnÄ… A^-1
np.linalg.inv(A) # pozwala odwracaÄ‡ tylko macierze kwadratowe
```

```python
np.linalg.pinv(A) # uogÃ³lniona metoda odwracania macierzy - 
                  #odwraca macierze niespeÅ‚niajÄ…ce koniecznych warunkÃ³w np. nie sÄ… kwadratowe
```

```python
np.linalg.pinv(B) #w mateatyce odwracanie niekwadratowej macierzy jest niedozwolone
```

```python
np.linalg.inv(A) @ A
```

```python
np.round(np.linalg.inv(A) @ A)
```

```python
np.round(np.abs(np.linalg.inv(A) @ A))
```

```python
#5 ZnajdÅº A^3
A @ A @ A
np.linalg.matrix_power(A,3)
```

```python
A**3 #podniesienie do 3 potÄ™gi wszystkich elementÃ³w macierzy to co innego niÅ¼ A*A*A
```
