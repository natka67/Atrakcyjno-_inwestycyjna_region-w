# Problem - chcemy zamodelować system ewidencyjny schorniska dla zwierząt, które przeprowadza informatyzację

```python
class Zwierz:
    """Klasa odpowiedzialna za ogólne cechy każdego zwierzaka, nie rozróżnia na gatunki, jest to klasa abstrakcyjna
    
    Atrybuty
    --------
    imie: string
        imie naszego zwierzaka
    wiek: int
        wiek zwierzaka w latach
    koszt_roczny: int
        roczny koszt utrzymania zwierzaka w schronisku
    glos: string
        dzwiek ktory wydaje zwierzak
    
    Metody
    ------
    zmien_imie()
        umozliwia zmiane imienia
    wypraw_urodziny()
        zwieksza wiek zwierzaka
    oblicz_koszt()
        pozwala na obliczenie kosztu zwierzaka w arbitrarnym czasie
    daj_glos()
        wypisuje dzwiek ktory wydaje zwierzak
    """
    
    def __init__(self, imie, wiek, koszt):
        self.imie = imie
        self.wiek = wiek
        self.koszt = koszt
        self.glos = "" #nieznane bo nie znamy gatunku zwierzaka!
        
    def zmien_imie(self, nowe_imie):
        self.imie = nowe_imie
        
    def wypraw_urodziny(self):
        self.wiek+=1
        
    #koszt powinien byc wyrażony w latach, np dla obliczenia kosztu utrzymania na 18 miesiecy, czas = 1.5
    def oblicz_koszt(self, czas):
        return self.koszt*czas
    
    def daj_glos(self):
        return self.imie + " mówi: " +self.glos
```

```python
azor = Zwierz("azor", 12, 200)
```

```python
azor.wypraw_urodziny()
```

```python
azor.wiek
```

```python
azor.daj_glos()
```

```python
class Kot(Zwierz):
    """Klasa pozwalajaca na modelowanie kota, dziedziczy po klasie Zwierz"""
    def __init__(self, imie, wiek, koszt):
        self.imie = imiet
        self.wiek = wiek
        self.koszt = koszt
        self.glos = "meow"
```

```python
jakis_kitku = Kot("szaroburek", 3, "50")
```

```python
jakis_kitku.daj_glos()
```

```python
jakis_kitku.glos
```
